```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt

# create a data set
X = np.sort(5 * np.random.rand(80, 1), axis = 0)
y = np.sin(X).ravel()
y[::5] += 0.5 * (0.5 - np.random.rand(16)) # gürültü ekliyoruz, nizami değerler olmasın

# plt.plot(X, y)
# plt.scatter(X, y) # boşluklu oluyor

# 2 farkli dt regresyon yapacagiz
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=9)
regr_1.fit(X, y) # training
regr_2.fit(X, y) # training

X_test = np.arange(0, 5, 0.05)[:, np.newaxis] # 0'dan 5'e kadar 0.05 artarak test verisi olusturuyoruz
y_pred_1 = regr_1.predict(X_test)
y_pred_2 = regr_2.predict(X_test)

# sonuclari gorsellestirelim
plt.figure()
plt.plot(X, y, c="red", label="data")
plt.scatter(X, y, c="red", label="data")
plt.plot(X_test, y_pred_1, color="blue", label="Max Depth: 2", linewidth=2)
plt.plot(X_test, y_pred_2, color="green", label="Max Depth: 9", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.legend()
```
