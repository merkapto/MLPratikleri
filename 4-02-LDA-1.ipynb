```python
from sklearn.datasets import fetch_openml
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt
import numpy as np

mnist = fetch_openml('mnist_784', version=1) #, as_frame=False
X, y = mnist["data"], mnist["target"] # X is (70000, 784), y is (70000,)
# print(X.shape, y.shape) # (70000, 784) (70000,)

lda = LinearDiscriminantAnalysis(n_components=2) # n_components must be < n_classes - 1

X_lda = lda.fit_transform(X, y) # X_lda is (70000, 2)
# print(X_lda.shape)  # (70000, 2)

# plt.figure(figsize=(10, 8))
plt.figure() # (figsize=(8, 6)) # Default size is usually fine for quick plots
# scatter = plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y.astype(int), cmap='tab10', alpha=0.7)
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y.astype(int), cmap='tab10', alpha=0.7)
# plt.colorbar(scatter, ticks=range(10), label='Digit Label')
plt.colorbar(label='Digit Label')
# plt.clim(-0.5, 9.5) # Set color limits for better colorbar ticks
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title('LDA of MNIST Dataset')
plt.grid(True)
plt.show()
```
