```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA

# Iris veri setini yukle
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

pca = PCA(n_components=2) 
X_pca = pca.fit_transform(X) # PCA sinif etiketlerini almaz
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X, y) # LDA sinif etiketlerini de alir

# PCA sonucununu gorsellestir
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
for i, target_name in enumerate(target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], label=target_name)
plt.title('PCA ile Boyut Indirme')
plt.xlabel('Birinci Bilesen')
plt.ylabel('Ikinci Bilesen')
plt.legend()
plt.grid()
# LDA sonucununu gorsellestir
plt.subplot(1, 2, 2)
for i, target_name in enumerate(target_names):
    plt.scatter(X_lda[y == i, 0], X_lda[y == i, 1], label=target_name)
plt.title('LDA ile Boyut Indirme')
plt.xlabel('Birinci Bilesen')
plt.ylabel('Ikinci Bilesen')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
